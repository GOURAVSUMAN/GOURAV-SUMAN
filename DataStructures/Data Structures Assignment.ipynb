```python
#Question(1):-

#Stack using two queue.

#STEPS:-

#define queue1 and queue2
#puch and pop operation and append the input.
#pop from queue1 and push from queue2 to queue1
class Stack:
    
    #In this i initializer a new instance of a class. 
    def __init__(self,queue1):
        #Create and initialize  two queue
        self.queue1=queue1
        self.queue2=[]
        
    #In this i add elemenyts to queue.    
    def add(self,data):
        #add data into queue.
        b=self.queue1.append(data)
        #print(self.queue1)
        return b
        
    #In this i pop elemenyts to queue.       
    def remove(self,data):
        #remove data from queue1 and add to queue2
        while(len(self.queue1))>0:
            a=self.queue1.pop()
            self.queue2.append(a)
        #Return the value and terminates the function.       
        return self.queue2

#Creating stack
s=Stack([])
#Print statement.
print(s.add([10,12]))
print(s.remove(1))
```
